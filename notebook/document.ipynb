```python
import os

os.makedirs("../data/text_files", exist_ok = True)
```

```python
sample_texts = {
    "../data/text_files/Ronaldo_intro.txt": "HE IS THE BULLY OF THE BIG BOYS... FEARFUL OF NOONE BETTER THAN ALMOST EVERYONE",
    "../data/text_files/Messi_intro.txt": "GENIUSSS ABSOLUTE GENIUS... ANOTHER ONE IN THE RECORD BOOKS."
}
```

```python
for file_path, content in sample_texts.items():
    with open(file_path, "w", encoding="utf-8") as f:
        f.write(content)
```

```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader("../data/text_files/Ronaldo_intro.txt", encoding="utf-8")
document = loader.load()
print(document)
# print(loader.load())
```

```python
from langchain_community.document_loaders import PyMuPDFLoader, PyPDFLoader
```

```python
from langchain_community.document_loaders import DirectoryLoader

dir_loader = DirectoryLoader(
    "../data/text_files",
    glob = "**/*.txt",
    loader_cls = TextLoader,
    loader_kwargs = {"encoding": "utf-8"},
    show_progress = False
)
documents = dir_loader.load()
print(documents)
```
